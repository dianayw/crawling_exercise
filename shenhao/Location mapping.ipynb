```python
import folium
from folium.plugins import MarkerCluster
from folium import plugins
import pandas as pd
```

```python
loaction_data = pd.read_csv('RtoSearchExport.csv')
```

```python
loaction_data.head()
```

```python
loaction_data.isna().sum()
```

```python
# choose the rows where Longitude and Latitude is not NaN
loaction_data = loaction_data[loaction_data['Longitude'].notna()]
loaction_data = loaction_data[loaction_data['Latitude'].notna()]
```

```python
# au latitude and longitude values
au_lati = -25.2744
au_longi = 133.7751

# define map of sydney and show
au_map = folium.Map(location=[au_lati,au_longi], zoom_star=12)

# get coordinates
coordinate_data = loaction_data.iloc[:,:]

# Instantiate a feature group for the coordinates in the dataframe
coordinate = folium.map.FeatureGroup()

# Loop through the coordinates and add each to the incidents feature group
for lat, lng, in zip(loaction_data.Latitude, loaction_data.Longitude):
    coordinate.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, 
            color='dodgerblue',
            fill=True,
            fill_color='white',
            fill_opacity=0.8
        )
    )

# Add coordinates to map
au_map = folium.Map(location=[au_lati, au_longi], zoom_start=4.3)
au_map.add_child(coordinate)
```

```python
# group points to have further understanding on location
au_map = folium.Map(location=[au_lati, au_longi], zoom_start=4.3)

# instantiate a mark cluster object for the coordinate in the dataframe
coordinate = plugins.MarkerCluster().add_to(au_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(loaction_data.Latitude, loaction_data.Longitude, loaction_data.Longitude):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(coordinate)

# add coordinate to map
au_map.add_child(coordinate)
```
